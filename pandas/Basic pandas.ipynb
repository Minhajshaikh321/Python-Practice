```python
import pandas as pd

# Define a dictionary containing students data
data = {'Name': ['Sneha', 'Shreya',
				'Sabhya', 'Riya'],
		'Age': [22, 18, 10, 19],
		'Stream': ['Computer', 'Commerce',
				'Arts', 'Mechanical'],
		'Percentage': [89, 93, 97, 73]}
df=pd.DataFrame(data,columns=['Name','Age','Stream','Percentage'])

print('Given data frame \n',df)
```

```python
for index,row in df.iterrows():
    print(index,row['Name'],row['Age'],row['Stream'],row['Percentage'])
```

```python
import pandas as pd

data = {'Name': ['Amit', 'Riya', 'Raj'], 'Age': [25, 28, 22]}
df = pd.DataFrame(data, index=['a', 'b', 'c'])
print(df)
print(df.loc['b', 'Age'])      # Access row with label 'b' and column 'Age'
print(df.loc['a':'c', ['Name']])  # Select multiple rows and a specific column
```

```python
print(df.iloc[1, 1])       # 1st row, 1st column (0-based index)
print(df.iloc[0:2, 0])     # Select first 2 rows, first column


#Use .loc[] when your data has meaningful row/column labels

#Use .iloc[] when you just want to access by numeric index
```
