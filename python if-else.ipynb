```python
                                                           if-elif
```

```python
#The elif keyword is Python's way of saying "if the previous conditions were not true, then try this condition".
a = 20
b = 20
if a > b:
    print("a is greater then b")
elif a == b :
    print("a is equal to b")
```

```python
a = 22
b = 300
if a > b:
    print("a is greter then b")
elif a == b :
    print("a is equal to b")
else:
    print("a is less then b")
```
