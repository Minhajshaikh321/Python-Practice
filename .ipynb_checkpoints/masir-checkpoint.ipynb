                                                             Recursion

```python
#recursion is a function which is calls itself it is used to directly 
#use a methematical formulaas a function 

'''
factorial(0) = 1
factorial(1) = 1
factorial(2) = 2*1*0
factorial(3) = 3*2*1*0
factorial(4) = 4*3*2*1*0
factorial(5) = 5*4*3*2*1*0

factorial(n) = n * factorial(n-1)
'''

def factorial(n):
    if(n == 1 or n == 0):
        return 1
    return n* factorial(n-1)

n = int(input("Ener a number :"))
print(f"the nfactorial of this number is{factorial(n)}")
```

                                                     16.Inheritance and more on oops

```python
#inheritence is a way of creating a new class from the existing class
                       #Single inheritence:base class to derived class
                       #Multiple inheritence:parent 1 class parent 1 class to chlid class
                       #Multilevel inheritence:parent class to chlid 1 or chlid 2

class Employee:
    company = "ITC ford" # bas class is chlid class
    name = "future"
    def show(self):
        print(f"the name of my company is {self.company} this is very usefull to our {self.name}")

class coder:       #base class is child class
    language="Java"
    def printlanguage(self):
        print(f"the {self.language} is very easy to read and write")

class programming(Employee,coder): #derived class is parent class
    company = "coding"
    def print(self):
        print(f"the name of {self.language} and he is good for {self.company}")


a = Employee()
b = programming()

b.show()
b.printlanguage()
b.print()
```

                                                       Multilevel inheritence

```python
class employee:
    a = 1

class programming(employee):
    b = 2

class manager(programming):
    c = 3

p = employee()
print(p.a)

p = programming()
print(p.a,p.b)

p = manager()
print(p.a,p.b,p.c)
```

                                                        super()methods

```python
class employee:
    def __init__(self):
        print("constructor of employee")
    a = 1

class programmming(employee):
    def __init__(self):
        print("constructor of programming")
    
    b = 2

class manager(programming):
    def __init__(self):
        super().__init__()
        print("constructor of manager")
    c = 3

p = employee()
print(p.a)

p = manager()
print(p.a,p.b,p.b)
```
