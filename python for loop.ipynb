                                                    python for loops

```python
python has two primitive command:
while loop
for loop
```

```python
#With the while loop we can execute a set of statements as long as a condition is true.
i = 0
while i < 6:
    print(i)
    i += 1
```

```python
#the break statement we can stop the loop even if the while condition is true:
i = 0
while i < 10:
    print(i)
    if i == 6:
     break
    i += 1
```

```python
#the continue statement we can stop the current iteration, and continue with the next:
i = 0
while i < 6:
    i += 1 
    if i == 3:
        continue
    print(i)
```

```python
#else using while loop
i = 0
while i < 5:
    print(i)
    i += 1   
else:
    print("i is nolonger then 6")
```

                                                       python for loop

```python
#A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

fruits = ["apple","mango","banana","cherry","pineapple"]
for i in fruits:
    print(i)
```

```python
#python looping through a string
x = "banana"
for i in x:
    print(i)
```

```python
#break and continue statement
fruits = ["apple","mango","banana","cherry","pineapple"]
for i in fruits:
    print(i)
    if i == "cherry":
        break

vegetables = ["apple","mango","pineapple"]
for x in vegetables:
    if x == 7:
        continue
    print(x)
```

```python
#the range is working with sequence functions
for i in range(6):
    print(i)
```

```python
for i in range(2,7):
    print(i)
```

```python
for i in range(2,20,3):#start,stop,step
    print(i)
```

```python
#else throuht afor loop
for i in range(5):
    print(i)
else:
    print("finally finished")
```

```python
#break statement in for loop
for i in range(5):
    if i == 4:break
    print(i)
else:
    print("finally finised")
```

```python
#A nested loop is a loop inside a loop.
adj = ["tasty","yummy","dilicious"]
fruits = ["apple","mango","kiwi"]
for x in adj:
   for y in fruits:
     print(x,y)
```
