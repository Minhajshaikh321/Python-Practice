# Current date in different format

```python
from datetime import date
today=date.today()

#d/m/y
d1=today.strftime("%d/%m/%Y")
print("d1",d1)

#textual month
d2=today.strftime("%B %d %Y")
print("d2",d2)

#month abbrevation day and year
d3=today.strftime("%b-%d-%Y")
print("d3",d3)
```

# datetime.timedelta Class

```python
#A timedelta object represents the difference between two dates or times. For example,
from datetime import datetime,date
t1=date(year=2018,month=7,day=12)
t2=date(year=2017,month=5,day=12)
t3=t1-t2
print(t3)
```

```python
#Difference between two timedelta objects

from datetime import timedelta

t1 = timedelta(weeks = 2, days = 5, hours = 1, seconds = 33)
t2 = timedelta(days = 4, hours = 11, minutes = 4, seconds = 54)

t3 = t1 - t2

print("t3 =", t3)
```

# strftime() method

```python
from datetime import datetime

time=datetime.now()

t=time.strftime("%H:%M:%S")
print('time',t)

t1=time.strftime("%m/%d/%Y, %H:%M:%S")
print("m/d/y hr/min/sec",t1)
```

# strptime() method

```python
#The strptime() method takes two arguments:1.a string representing date and time 2.format code equivalent to the first argument

from datetime import datetime

date_str="28 December 2001"

obj=datetime.strptime(date_str,"%d %B %Y")
print('obj',obj)
```

# Create a digit clock in python

```python
import time
while True:
    localtime=time.localtime()
    res=time.strftime("%I:%M:%S %p",localtime)
    print(res)
    time.sleep(1)
   
```
