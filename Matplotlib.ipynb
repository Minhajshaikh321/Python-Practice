```python
import pandas as pds
import matplotlib.pyplot as plt
import itertools

final=[{'qn':'aaaa','10': 1, '20': 3, '30': 3, '40': 1}, 
       {'qn':'bb','lotus school': 3, 'st martin school': 3, 'Orchid school': 1, 'msi school': 1}, 
       {'qn':'cccc','70%': 3, '60%': 2, '80%': 3}, 
       {'qn':'ddd','Yes': 6, 'No': 2}]
for i in final:
    slice = dict(itertools.islice(i.items(), 1 ,None))
    v=slice.values()
    series=pds.Series(v,slice.keys())
    series.plot.pie(label='',title='show data',autopct='%2.1d%%');
    plt.legend()
    plt.show(block=True)
```
