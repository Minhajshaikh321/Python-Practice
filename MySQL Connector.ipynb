```python
import mysql.connector

try:
    connection = mysql.connector.connect(host='localhost',database='sql_store',user='root',password='Minhaj123')
    sql_select_Query = """select * from customers"""
    cursor = connection.cursor()
    cursor.execute(sql_select_Query)
    records = cursor.fetchall()
    print('total no.of rows=',cursor.rowcount)
    print('\nprint each row')
    for row in records:
        print('Customerid= ',row[0],)
        print('firstname= ',row[1],)
        print('lastname= ',row[2],)
        print('birthdate= ',row[3],)
        print('phone= ',row[4],) 
        print('address= ',row[5])
        print('city= ',row[6])
except mysql.connector.Error as e:
    print("Error reading data from my_table", e)
finally:
    if connection.is_connected():
        connection.close()
        cursor.close()
        print("MySQL connection is closed")
```

```python
import mysql.connector

try:
    connection = mysql.connector.connect(host='localhost',database='minhajdb',user='root',password='Minhaj123')
    sql_query="""select * from employee"""
    cursor=connection.cursor()
    cursor.execute(sql_query)
    records = cursor.fetchall()
    print('total no.of rows=',cursor.rowcount)
    print('\nprint each row')
    for row in records:
        print(row[0],row[1],row[2],row[3])
        print("id= ",row[0])
        print("employee name= ",row[1])
        print("department= ",row[2])
        print("salary= ",row[3])
except mysql.connector.Error as e:
    print("Error reading data from my_table", e)
finally:
    if connection.is_connected():
        connection.close()
        cursor.close()
        print("MySQL connection is closed")
```
