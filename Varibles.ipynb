```python
#varibles namesare case-sensitive
myvar = "john"
my_var = "john"
MYVAR = "john"
_my_var = "john"
myvar908 = "john"
print(myvar)
print(my_var)
print(MYVAR)
print(_my_var)
print(myvar908)
#illigal variables(var name,2varname,var-name)

#Many Values to Multiple Variables
x = "orange"
y = "pineapple"
z = "cherry"
print(x)
print(y)
print(z)

#One Value to Multiple Variables
x = y = z = "orange"
print(x)
print(y)
print(z)

#unpack a collection

fruits = "Amaan","Nabil","Rehan"
(x,y,z) = fruits
print(x)
print(y)
print(z)
```

```python
#Global Variables
x = "awesome"

def func():
    print("python is " + x)

func()
#--------------------------------------#
x = "Easy to learn"
def myfunc():
    x = "Fantastic"
    print("python is "+ x)

myfunc()

print("python is " + x)    
```

```python
#Global keyword
def myfunc():
    global x
    x = "fantastic"

myfunc()
print("python is " + x)

x = "awesome"

def myfunc():
    global x
    x = "Good"
    print("python is " + x)

myfunc()
```

```python
sum = 0
counter = 0
numbers = [1, 2, 3, 4]
while counter < len(numbers):
  sum = sum + numbers[counter]
  counter = counter + 1
print(sum)
```
