```python
#The union() method returns a new set with all items from both sets.
set1 = {"apple","mango","cherry"}
set2 = {"potatao","tomato"}

set3 = set1.union(set2)
print(set3)

#You can use the | operator instead of the union() method, and you will get the same result.

set1 = {"apple","mango","pineapple"}
set2 = {"kiwi","cherry"}
set3 = set1 | set2
print(set3)

#jions multiple sets
set1 = {"apple","mango","cherry"}
set2 = {"kiwi","pineapple"}
set3 = {"tomato","potato"}
set4 = {23,29,35,15}

set5 = set1 | set2 | set3 | set4 #jioning the sets
print(set5)
```

```python
#The update() changes the original set, and does not return a new set.
set1 = {"apple","mango","kiwi"}
set2 = {"banana","watermelon","orange"}
set2.update(set1)
print(set2)
```

```python
#The intersection() method will return a new set, that only contains the items that are present in both sets.
set1 = {"apple","mango","cherry","watermelon","melon"}
set2 = {"pineapple","apple","melon","grapes","banana"}
set3 = set1.intersection(set2)
print(set3)

#You can use the & operator instead of the intersection()
set1 = {"apple","mango","grapes"}
set2 = {"google","microsoft","apple"}
set3 = set1 & set2
print(set3)
```

```python
#difference() Keep all items from set1 that are not in set2:
set1 = {"apple","mango","pineapple"}
set2 = {"potato","tomato","apple"}

set3 = set1.difference(set2)
print(set3)

#You can use the - operator instead of the difference() method, and you will get the same result.
set1 = {89,34,1,23}
set2 = {90,91,1,23}

set3 = set1 - set2
print(set3)
```

```python
#symmetric_difference() method will keep only the elements that are NOT present in both sets.

set1 = {"apple","mango","cherry","banana"} #'apple' intersection is not print in symmetric_difference
set2 = {"microsoft","google","apple","firefox"} #'apple'

set3 = set1.symmetric_difference(set2)
print(set3)

set1 = {"apple","mango","orange"}
set2 = {"microsoft","google","apple"}
set3 = set1 ^ set2
print(set3)
```

```python
def test1():  # defining 1st function
    price = 900  # local variable
    print("Value of price in test1 function :", price)

def test2():  # defining 2nd function
    # NameError: name 'price' is not defined
    print("Value price in test2 function:", price)

# call functions
test1()
test2()
```

```python
def test1():  # defining 1st function
    global price
    price = 900  # local variable
    print("Value of price in test1 function :", price)

def test2():  # defining 2nd function
    # NameError: name 'price' is not defined
    print("Value price in test2 function:", price)

# call functions
test1()
test2()
print('price is',price)
```
