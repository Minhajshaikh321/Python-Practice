# Method1:USing List

```python
stack=[]

#push element
stack.append(10)
stack.append(20)
stack.append(30)
print(f'Stacks after pushes {stack}')

#pop element
pop_item=stack.pop()
print(f'Pop item {pop_item}')

# Peek top item
print("Top element:", stack[-1])

# Check if empty
print("Is stack empty?", len(stack) == 0)
```

# Method 2: Using collections.deque (Faster & Efficient)

```python
from collections import deque

stack=deque()

stack.append('A')
stack.append('B')
stack.append('C')

print(f'Stacks {stack}')
print("Popped:", stack.pop())
print("After pop:", stack)
```

# Method 3: Using a Class (Custom Stack Implementation)

```python
class Stack:
    def __init__(self):
        self.stack = []
    def push(self,item):
        self.stack.append(item)
    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        return "Stack is empty"
    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        return "Stack is empty"
    def is_empty(self):
        return len(self.stack) == 0
    def size(self):
        return len(self.stack)
    def get_stack(self):
        if not self.is_empty():
            return self.stack
        return "Stack is empty"

s = Stack()
s.push(10)
s.push(20)
s.push(30)
print("Stack :", s.get_stack())
print("Stack top:", s.peek())
print("Popped item:", s.pop())
print("Stack size:", s.size())
print("Stack :", s.get_stack())
```
