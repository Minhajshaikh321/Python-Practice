# A HashMap (also known as a Hash Table) is a key–value data structure that allows fast access, insertion, and deletion — usually in O(1) time.

```python
#a HashMap is implemented as a dictionary (dict).
student = {
    "name": "Zohran",
    "age": 34,
    "city": "Newyork"
}
print(student["name"])  # Access by key
```

```python
Key   → Hash → Index → Value
'name' → 1123 → 3 → 'Zohran'
```

# basic hashmap usage

```python
employee = {"E101": "John","E102": "Sara","E103": "Ali"}

employee["E104"]="Sam"  #add new k-v pair

print('Access a value ',employee["E103"])  #access value

employee.pop("E101")  #remove key

for key, value in employee.items():
    print(key, "→", value)
```
