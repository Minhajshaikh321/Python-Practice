```python
def mergeSort(arr):
  if len(arr) <= 1:
    return arr

  mid = len(arr) // 2
  leftHalf = arr[:mid]
  rightHalf = arr[mid:]
  print(f'left half {leftHalf}')
  print(f'right half {rightHalf}')


  sortedLeft = mergeSort(leftHalf)
  sortedRight = mergeSort(rightHalf)
  return merge(sortedLeft, sortedRight)

def merge(left, right):
  print(f'merge func left {left}')
  print(f'merge func right {right}')

  result = []
  i = j = 0

  while i < len(left) and j < len(right):
    print('left i in merge fun ',left[i],'right in merge func ',right[i])
    if left[i] < right[j]:
      result.append(left[i])
      i += 1
    else:
      result.append(right[j])
      j += 1

  result.extend(left[i:])
  result.extend(right[j:])
  print('restult of merge sort ',result)
  return result

mylist = [3, 7, 6, -10, 15, 23.5, 55, -13]
mysortedlist = mergeSort(mylist)
print("Sorted array:", mysortedlist)
```

```python
class MyMeta(type):
  def __new__(cls,name,bases,attrs):
    attrs['new_attrs']='Hello from meta class attr'
    return super().__new__(cls,name,bases,attrs)
class MyClass(metaclass=MyMeta):
  pass
print(MyClass.new_attrs)
```
