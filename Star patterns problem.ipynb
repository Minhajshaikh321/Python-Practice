# Increasing Triangle Pattern

```python
n=5
for i in range(n):
    for j in range(i+1):
        print('*',end=" ")
    print()    
```

# Decreasing Triangle Pattern

```python
n=5
for i in range(n):
    for j in range(i,n):
        print('*',end=" ")
    print()    
        
```

# Increasing space star Triangle Pattern

```python
n=5
for i in range(n):
    for j in range(i,n):
        print(' ',end=" ")
    for k in range(i+1):
        print('*',end=" ")
    print()    
```

# Decreasing space star Triangle Pattern

```python
n=5
for i in range(n):
    for j in range(i+1):
        print(' ',end=" ")
    for k in range(i,n):
        print('*',end=" ")
    print()    
```

# Pyramid Star Pattern

```python
n=5
for i in range(n):
    for j in range(i,n):
        print(' ',end=" ")
    for k in range(i):
        print('*',end=" ")
    for l in range(i+1):
        print('*',end=" ")
        
    print()    
```

```python
def print_pyramid(rows):
    for i in range(rows):
        print(' ' * (rows - i - 1) + '*' * (2 * i + 1))

rows = int(input("Enter the number of rows for the pyramid: "))
print_pyramid(rows)
```

```python
n=5
p=65
for i in range(n):
    for j in range(i+1):
        print(chr(p),end=" ")
    p+=1
    print()
```

```python
n=5
for i in range(n):
    p=65
    for j in range(i+1):
        print(chr(p),end=" ")
        p+=1
    print()    
```

```python
n=5
for i in range(n):
    p=65
    for j in range(i,n):
        print(" ",end=" ")
    for j in range(i):
        print(chr(p),end=" ")
        p+=1
    for j in range(i+1):
        print(chr(p),end=" ")
        p+=1
    print()    
```

```python
n=5
for i in range(n):
    p=65
    for j in range(i,n):
        print(' ',end=" ")
    for j in range(i):
        print(chr(p),end=" ")
        p+=1
    for j in range(i+1):
        print(chr(p),end=" ")
        p-=1
    print()    
```
